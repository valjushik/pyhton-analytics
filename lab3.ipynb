### Создание нового признака delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. (Если реновации дома не было, то в новом признаке 0)

```python
df['delta_renovated'] = (df['yr_renovated']-df['yr_built']).apply(lambda x: x if x > 0 else 0)
df[['yr_built', 'yr_renovated', 'delta_renovated']].sample(10)
```

### Создание признаков года продажи и месяца продажи

```python
df['year_sale'] = df['date'].apply(lambda x: int(x[:4]))
df['month_sale'] = df['date'].apply(lambda x: int(x[4:6]))
df[['id', 'date', 'year_sale', 'month_sale']].sample(10)
```

### Удаление date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```

## Задание 3

### Нахождени средней стоимости домов в зависимости от количества спален. Сортировка от меньшей стоимости к большей

```python
group = df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
```

### Нахожднение минимальной, средней и максимальной стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

### Построение таблицы с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index = 'waterfront', columns = 'view', values = 'id', aggfunc = 'count', fill_value = 0)
```

### Нахождение каких домов в зависимости от этажности и количества спален больше

```python
df.pivot_table(index = 'floors', columns = 'bedrooms', aggfunc = 'count', values = 'id', fill_value=0)
```

### Построение таблицы с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома
